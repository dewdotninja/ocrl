# ภาคผนวก A รวมนิยาม/สัญกรณ์/ศัพท์ และคณิตศาสตร์

ในภาคผนวกนี้เป็นการรวบรวมนิยาม สัญกรณ์ คำศัพท์ และหลักการพื้นฐานทางคณิตศาตร์ ที่เกี่ยวข้องกับเนื้อหาหลักของหนังสือ 
แต่มีรายละเอียดมากเกินกว่าจะใส่ไว้ในตัวบท

## A.1 ความเป็นคอนเวกซ์

จากเนื้อหาในหนังสือจะเห็นได้ว่าความเป็นคอนเวกซ์ (convexity) มีความสำคัญต่อการหาค่าเหมาะที่สุดเป็นอย่างมาก 
เพราะสำหรับปัญหาที่เป็นคอนเวกซ์ อัลกอริทึมหาค่าหมาะที่สุดจะลู่เข้าได้เร็วทำให้สามารถทำงานในอัตราการสุ่มที่สูงได้ 
ดังนั้นในภาคผนวกนี้จะลงรายละเอียดเพิ่มเติมโดยสังเขป

### A.1.1 เซตคอนเวกซ์

นิยามของเซตคอนเวกซ์ (convex set) ทางคณิตศาสตร์คือ เมื่อกำหนด $\mathbf{S}$ เป็นปริภูมิเวกเตอร์ (vector space) หรือปริภูมิสัมพรรค (affine space) บนเลขจำนวนจริง หรือโดยทั่วไปเป็นฟิลด์อันดับ (ordered field) เช่นปริภูมิยูคลิด (Euclidean spaces) เซตย่อย $\mathbf{C}$ ของ $\mathbf{S}$ เรียกว่าเป็นแบบคอนเวกซ์ ถ้า
$$
(1-\alpha)x + \alpha y, \; \forall x,y \in \mathbf{C}, \; \alpha \in [0,1] \tag{A.1} 
$$
เข้าใจนิยามนี้ได้ง่ายจากรูปทรงทางเรขาคณิตดังในรูปที่ A.1 คือเมื่อเลือก 2 จุดใดๆ ในเซต เส้นตรงที่เชื่อมระหว่าง 2 จุดจะต้องอยู่ในเซตทั้งหมด จากรูป เซต A, B, D เป็นแบบคอนเวกซ์ แต่ C,E ไม่ใช่เพราะมีบางส่วนของเส้นตรงที่อยู่ภายนอกเซต

ตัวอย่างของเซตคอนเวกซ์

* ปริภูมิย่อยเชิงเส้น : $Ax = b$
* ปริภูมิแบ่งครึ่ง/กล่อง/โพลีโทพ (polytopes) : $Ax \le b$
* ทรงกลม/รี (spheres/ellipsoids) : $x^TPx \le 1, \; P \succ 0$
* กรวยอันดับสอง (second-order cones) : $x_1 \ge \lVert x_{2:N} \rVert_2$ (หรือนอร์มแบบอื่นก็ได้ กรณีนอร์ม 2 จะเป็นกรวยรูปร่างเหมือนโคนไอศครีม)

### A.1.2 ฟังก์ชันคอนเวกซ์

นิยามของฟังก์ชันคอนเวกซ์มีหลายรูปแบบ ที่พบบ่อยคือ เมื่อกำหนด $\mathbf{X}$ เป็นเซตย่อยแบบคอนเวกซ์ของปริภูมิเวกเตอร์ $f(x) : \mathbb{R}^n \rightarrow \mathbb{R}$ 
เรียกว่าเป็นฟังก์ชันคอนเวกซ์ถ้าและเพียงแต่ถ้า 

$$
f(\alpha x_1 + (1 - \alpha) x_2) \le \alpha f(x_1) + (1 - \alpha) f(x_2) \;\; \forall x_1,x_2 \in \mathbf{X}, \; \alpha \in [0,1] \tag{A.2}
$$

หากพิจารณาจากรูปกราฟ ฟังก์ชันคอนเวกซ์มีคุณสมบัติคือที่ส่วนที่เหนือจากกราฟทั้งหมด (epigraph) เป็นเซตคอนเวกซ์ หรือเมื่อเลือก 2 จุดใดๆ บนกราฟ $f(x)$ เส้นตรงที่เชื่อมระหว่าง 2 จุดจะต้องไม่มีส่วนใดอยู่ใต้กราฟ ดังแสดงในรูปที่ A.2 ด้านซ้ายคือฟังก์ชันคอนเวกซ์ ด้านขวาคือไม่ใช่คอนเวกซ์ 

ตัวอย่างของฟังก์ชันคอนเวกซ์ 

* ฟังก์ชันเชิงเส้น : $f(x) = c^Tx$
* ฟังก์ชันกำลังสอง : $f(x) = \frac{1}{2}x^TQx + q^Tx, \;\; Q \succcurlyeq 0$
* นอร์มทุกประเภท : $f(x) = \lVert x \rVert_p, \;\; p = 1,2,\ldots, \infty$

รูปที่ A.2 การตรวจสอบฟังก์ชันคอนเวกซ์

ดังนั้นปัญหาการหาค่าเหมาะที่สุดแบบคอนเวกซ์คือ หาค่าต่ำสุดของฟังก์ชันคอนเวกซ์ $f(x)$ บนเงื่อนไขบังคับ $c(x)$ ที่เป็นเซตคอนเวกซ์ ตัวอย่างที่พบบ่อยเช่น

* การโปรแกรมเชิงเส้น (LP : linear program) : $f(x), c(x)$ เป็นแบบเชิงเส้นทั้งหมด
* การโปรแกรมกำลังสอง (QP : quadratic program) : $f(x)$ เป็นกำลังสอง $c(x)$ เป็นเชิงเส้น
* การโปรแกรมกำลังสองบนเงื่อนไขกำลังสอง (QCQP : quadratic-constrained quadratic program) : $f(x)$ เป็นกำลังสอง $c(x)$ เป็นกำลังสอง (หรือเรียกว่าเป็นเซตแบบทรงรี)
* การโปรแกรมแบบกรวยกำลังสอง (SOCP : second-order cone program) : $f(x)$ เป็นเชิงเส้น $c(x)$ เป็นแบบกรวย

**หมายเหตุ :** ในอดีตคำว่า "โปรแกรม" ในวิธีการต่างๆ ไม่จำเป็นต้องหมายถึงการเขียนโค้ดบนคอมพิวเตอร์ แต่แทนการหาค่าเหมาะที่สุดหรือการแก้ปัญหาเพื่อหาคำตอบแบบมีขั้นตอน 

### A.1.3 จุดเด่นของการหาค่าเหมาะที่สุดแบบคอนเวกซ์

ในปัจจุบันการหาคำตอบเหมาะที่สุดสำหรับปัญหาคอนเวกซ์ถูกประยุกต์ใช้ในงานหลายประเภท 
โดยเฉพาะงานสำคัญที่เกี่ยวข้องกับชีวิตประจำวันเช่นตัวควบคุม MPC ในอุปกรณ์รถยนต์ ดังนั้นความเชื่อถือได้ของตัวหาคำตอบเป็นสิ่งสำคัญยิ่งยวด 
เพราะอุปกรณ์จะต้องสามารถหาคำตอบได้อย่างรวดเร็วในทุกรอบของการทำงาน และต้องไม่มีการผิดพลาดแม้ครั้งเดียว เพราะหากมีความผิดพลาดเกิดขึ้น 
ผลิตภัณฑ์จำนวนมากอาจต้องถูกเรียกกลับซึ่งก่อความเสียหายอย่างมหาศาล ไม่นับถึงกรณีที่เกิดอุบัติเหตุสูญเสียต่อชีวิตและทรัพย์สิน 
หากปัญหาไม่เป็นแบบคอนเวกซ์ เราไม่สามารถรับประกันสิ่งเหล่านี้ได้เลย 

สรุปข้อดีของปัญหาการหาค่าเหมาะที่สุดแบบคอนเวกซ์

* เมื่อได้คำตอบที่ตรงตามเงื่อนไข KKT แน่ใจได้ว่าเป็นคำตอบวงกว้าง เพราะจะไม่มีทางเกิดคำตอบเฉพาะที่เหมือนฟังก์ชันทั่วไปที่ไม่เป็นคอนเวกซ์
* ในทางปฏิบัติ หากปรับแต่งอย่างเหมาะสม วิธีนิวตันลู่เข้าเร็วและเชื่อถือได้ (ไม่ควรเกิน 5-10 การวนซ้ำ) 
ดังนั้นสามารถคำนวณด้วยอัตราควบคุมที่สูงเพียงพอ
